```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read the dataset

data = pd.read_csv('accident_data.csv')
data.head()
```

```python
features = ['Start_Time','Temperature(F)','Wind_Chill(F)','Humidity(%)','Pressure(in)','Visibility(mi)','Precipitation(in)','Crossing','Junction','Station','Traffic_Signal','Severity']
data = data[features]
data.head()
```

```python
def percent_na(feature):
    na_val = 0
    for val in feature:
        if pd.isna(val):
            na_val += 1
    return na_val/(len(feature))

null_features = []
for feature in features:
    print("%s: %5.4f" % (feature, percent_na(data[feature])))
    if percent_na(data[feature]) > 0:
        null_features.append(feature)
```

```python
null_features
```

```python
def handle_na(feature):
    repl_val = np.mean(data[feature])
    data[feature] = data[feature].replace(np.nan, value = repl_val)
    print("Null values handled for %s" %(feature))
    return

for feature in null_features:
    handle_na(feature)
```

```python
data
```

```python
import datetime
from datetime import datetime

def day(x):
    return x.dayofweek

data['Start_Time'] = pd.to_datetime(data['Start_Time'])
data['Day'] = data['Start_Time'].apply(day)
data['Start_Time'] = data['Start_Time'].apply(lambda x: x.time().hour)
```

```python
data.head()
```

```python
def set_sev(x):
    if x == 1:
        return "low risk"
    elif x == 2:
        return "medium risk"
    return "high risk"

data['Risk'] = data['Severity'].apply(set_sev)
```

```python
data.head()
```

```python
def tod(x):
    '''
    
    Early Morning(6-9): 0
    Morning(9-12): 1
    Afternoon(12-16): 2
    Evening(16-20): 3
    Night(20-6): 4
    
    '''
    if x in range(6,9):
        return 0
    elif x in range(9,12):
        return 1
    elif x in range(12,16):
        return 2
    elif x in range(16,20):
        return 3
    return 4

data.rename(columns = {'Start_Time':'Time_of_Day'}, inplace = True)
data['Time_of_Day'] = data['Time_of_Day'].apply(tod)
data.head()
```

```python
data.describe()
```

```python
data = data.drop(['Severity'], axis = 1)
data
```

```python
from imblearn.over_sampling import SMOTE

X=data.drop(['Risk'], axis = 1)
y=data['Risk']

smote = SMOTE(random_state = 42)
X,y = smote.fit_resample(X,y)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split as tts

X_train,X_test,y_train,y_test = tts(X,y,test_size = 0.4)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators = 100, max_samples = 0.5, n_jobs = -1)
clf.fit(X_train, y_train)
```
